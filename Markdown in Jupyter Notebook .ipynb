# 6. Mathematical Symbols / Latex Maths
Use this code $ $ : $Mathematical Symbols$\
Jupyter Notebooks' Markdown cells support **LateX** for formatting mathematical equations. To tell Markdown to interpret your text as **Latex**, surround your input with *dollar signs* like this:

$z=\dfrac{2x}{3y}$

An equation can be very complex:

$F(k) = \int_{-\infty}^{\infty} f(x) e^{2\pi i k} dx$

If you want your *Latex equations to be indented towards the center of the cell*, surround your input with *two dollar signs* on each side like this:

$$2x+3y=z$$


**Mathematical signs**

1. (+) = addition
2. (-) = substraction
3. (/) = divide
4. (*) = multiplication
5. (**) = power



    

# 7. Indented quotes
Use a great than sign(>) and then a space, then type a text. The text is indented and has a grey horizontal line to the left.
> This is first line to 1st indented quotes \
This is second line to 1st quotes

> This is first line to 2nd indented quotes <br>
This is second line to 2nd quotes

>> This is first line to 2nd indented quotes <br>
This is second line to 2nd quotes

>>> This is first line to 2nd indented quotes <br>
This is second line to 2nd quotes

You will see #\ sign and #<br> code in above quotes, it works the same.
